```python
import pandas as pd
import numpy as np
```

```python
# reading the data from the Excel file
harvest = pd.read_excel('maize.xlsx', sheet_name='harvest')
land_use = pd.read_excel('maize.xlsx', sheet_name='land_use')
fertilizer = pd.read_excel('maize.xlsx', sheet_name='fertilizer')
maize_land = pd.read_excel('maize.xlsx', sheet_name='maize_land')
mixed = pd.read_excel('maize.xlsx', sheet_name='mixed')
```

```python
# checking for missing values
harvest.isnull().sum()
```

```python
land_use.isnull().sum()
```

```python
fertilizer.isnull().sum()
```

```python
# merging the dataframes
merged_data = pd.merge(harvest, land_use, on='District', how='inner')
merged_data = pd.merge(merged_data, fertilizer, on='District', how='inner')
merged_data = pd.merge(merged_data, maize_land, on='District', how='inner')
merged_data = pd.merge(merged_data, mixed, on='District', how='inner')
```

```python
# checking missed new data set 
merged_data.info()
```

```python
merged_data = merged_data.fillna(0)
```

```python
import matplotlib.pyplot as plt

# Which district has the highest maize production
plt.bar(merged_data['District'], merged_data[''])    
plt.xlabel('District')
plt.ylabel('Maize Harvest (tons)')
plt.title('Maize Harvest by District')
plt.xticks(rotation=90)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression as lr
```

```python
# Machine Learning Model training
agri_model = lr()
X = merged_data[['cultivated_land', 'Inorganic fertilizer', 'Organic fertilizer', 'Modern irrigated agricultural land (Ha)']]
y = merged_data['harvest']
agri_model.fit(X, y)
```

```python
# model testing
agri_model.predict(X)
```

```python
agri_model.score(X, y, sample_weight=None)
```
